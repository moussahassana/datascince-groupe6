Dans le contexte des Convolutional Neural Networks (CNN), la convolution est une opération essentielle pour extraire des caractéristiques significatives des données d'entrée, généralement des images. Voici comment cela fonctionne :

Image d'entrée: Tout d'abord, une image est présentée au réseau CNN. Cette image est généralement représentée sous forme de matrice de valeurs de pixels, où chaque valeur représente l'intensité lumineuse d'un pixel spécifique.

Filtres ou noyaux: Les CNN utilisent des filtres, également appelés noyaux, pour analyser l'image. Ces filtres sont de petites matrices de nombres qui sont appliquées à des régions spécifiques de l'image. Chaque filtre est conçu pour détecter un certain motif ou une caractéristique, comme des bords, des textures ou des motifs.

Opération de convolution: Pour chaque région de l'image, le filtre est appliqué en effectuant une opération appelée convolution. Cela consiste à superposer le filtre sur la région de l'image, puis à multiplier chaque élément de la région de l'image par le filtre correspondant, et à sommer les résultats. Cette opération est répétée pour chaque région de l'image.

Carte des caractéristiques: Lorsque le filtre est déplacé sur toute l'image, une nouvelle matrice appelée carte des caractéristiques est créée. Cette carte des caractéristiques met en évidence les régions de l'image où le motif ou la caractéristique spécifique détecté par le filtre est présent.

Opérations supplémentaires: Après la convolution, d'autres opérations peuvent être appliquées dans un CNN, telles que l'activation avec une fonction d'activation non linéaire (comme ReLU), le sous-échantillonnage (pooling) pour réduire la dimensionnalité, ou encore la superposition de plusieurs couches de convolution pour extraire des caractéristiques de plus haut niveau.

En résumé, la convolution dans un CNN permet de détecter et d'extraire des motifs visuels significatifs à différentes échelles spatiales, formant ainsi la base de la capacité du réseau à comprendre et à interpréter des images pour des tâches telles que la classification, la détection d'objets ou la segmentation sémantique.

#### 5.	Backward Pass : Passage en arrière dans les CNN 

Le passage en arrière dans les CNN consiste à calculer les gradients de la fonction de perte par rapport aux poids (ou filtres) et par rapport au biais de la couche de convolution. Ces gradients sont ensuite utilisés pour mettre à jour les poids et les biais pendant le processus d'apprentissage.

Dérivée de la perte par rapport aux poids/filtres :
La dérivée de la perte par rapport à un poids ou un filtre   est calculée en multipliant la dérivée de la perte par rapport à la sortie de la couche    par la dérivée de la sortie par rapport au poids  , qui est équivalente à l'entrée   à cette position.



<b>Mise à jour des filtres :</b>
Une fois que la dérivée de la perte par rapport à un poids est calculée, les filtres sont mis à jour en soustrayant une fraction de cette dérivée, multipliée par le taux d'apprentissage  , de l'ancienne valeur du filtre.



<b>Mise à jour du biais :</b>
De manière similaire, le biais est mis à jour en soustrayant une fraction de la dérivée de la perte par rapport au biais, multipliée par le taux d'apprentissage  de l'ancienne valeur du biais.

### V Démontration  

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
from ann_visualizer.visualize import ann_viz
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

# On charge les données d'entraînement et de test
(x_train_full, y_train_full), (x_test, y_test) = mnist.load_data()

# On Utilise seulement une partie du dataset pour des raisons de performance, c'est pas facile
x_train, y_train = x_train_full[:30000], y_train_full[:30000]

# Prétraitement des données
x_train = x_train.reshape((len(x_train), 28, 28, 1)) / 255.0
x_test = x_test.reshape((len(x_test), 28, 28, 1)) / 255.0

# Construction du modèle CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compilation du modèle
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entraînement du modèle
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.2)

# Affichage de la courbe d'évaluation de l'accuracy
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

# Évaluation sur la moitié du dataset de test
x_test_half, y_test_half = x_test[:len(x_test)//2], y_test[:len(y_test)//2]
test_loss, test_acc = model.evaluate(x_test_half, y_test_half)
print(f"Test accuracy: {test_acc}")

# Prédiction sur le dataset de test
y_pred = np.argmax(model.predict(x_test_half), axis=-1)

# Affichage de la matrice de confusion
conf_matrix = confusion_matrix(y_test_half, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion matrix')
plt.show()

# Inférence sur une seule image
sample_index = 0
sample_image = x_test_half[sample_index]
sample_label = y_test_half[sample_index]
sample_prediction = np.argmax(model.predict(np.expand_dims(sample_image, axis=0)))
print(f"True label: {sample_label}")
print(f"Predicted label: {sample_prediction}")
plt.imshow(sample_image.squeeze(), cmap='gray')
plt.show()
ann_viz(model, title="CNN_Model_Architecture", view=True)
```
